```python
# ننشئ مخططين فرعيين (subplots) جنب بعض لمقارنة أطوال المراجعات الإيجابية والسلبية
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 8))

# المخطط الأول: توزيع أطوال المراجعات الإيجابية
ax1.hist(df[df['sentiment'] == 'positive']['review'].str.len())
ax1.set_title('Positive Reviews')  # عنوان المخطط الأول

# المخطط الثاني: توزيع أطوال المراجعات السلبية
ax2.hist(df[df['sentiment'] == 'negative']['review'].str.len())
ax2.set_title('Negative Reviews')  # عنوان المخطط الثاني
```

```python
# نجمع كل النصوص الخاصة بالمراجعات الإيجابية في نص واحد
text = " ".join(i for i in df[df['sentiment'] == 'positive']['review'])

# ننشئ سحابة كلمات (wordcloud) للنصوص الإيجابية مع خلفية بيضاء
wordcloud = WordCloud(background_color="white").generate(text)

# نعرض سحابة الكلمات في مخطط
plt.figure(figsize=(15, 10))  # تحديد حجم الشكل
plt.imshow(wordcloud, interpolation='bilinear')  # عرض سحابة الكلمات
plt.axis("off")  # إخفاء المحاور
plt.title('Wordcloud for Positive Review')  # عنوان المخطط
plt.show()  # إظهار المخطط
```

```python
# نجمع كل النصوص الخاصة بالمراجعات السلبية في نص واحد
text = " ".join(i for i in df[df['sentiment'] == 'negative']['review'])

# ننشئ سحابة كلمات (wordcloud) للنصوص السلبية مع خلفية بيضاء

# سحابة الكلمات بدون إزالة الكلمات الشائعة
wordcloud = WordCloud(background_color="white").generate(text)

# نعرض سحابة الكلمات في مخطط
plt.figure(figsize=(15, 10))  # تحديد حجم الشكل
plt.imshow(wordcloud, interpolation='bilinear')  # عرض سحابة الكلمات
plt.axis("off")  # إخفاء المحاور
plt.title('Wordcloud for Negative Review')  # عنوان المخطط
plt.show()  # إظهار المخطط
```

```python
# نغير اسم العمود من 'review' إلى 'text' 
df.rename(columns={'review': 'text'}, inplace=True)
df
```

```python
def cleaning(text):        
    # converting to lowercase, removing URL links, special characters, punctuations...
    text = text.lower() # converting to lowercase
    text = re.sub('https?://\S+|www\.\S+', '', text) # removing URL links
    text = re.sub(r"\b\d+\b", "", text) # removing number 
    text = re.sub('<.*?>+', '', text) # removing special characters, 
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text) # punctuations
    text = re.sub('\n', '', text)
    text = re.sub('[’“”…]', '', text)
   
    #removing emoji: 
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    text = emoji_pattern.sub(r'', text)   

   # removing short form: 
    
    text=re.sub("isn't",'is not',text)
    text=re.sub("he's",'he is',text)
    text=re.sub("wasn't",'was not',text)
    text=re.sub("there's",'there is',text)
    text=re.sub("couldn't",'could not',text)
    text=re.sub("won't",'will not',text)
    text=re.sub("they're",'they are',text)
    text=re.sub("she's",'she is',text)
    text=re.sub("There's",'there is',text)
    text=re.sub("wouldn't",'would not',text)
    text=re.sub("haven't",'have not',text)
    text=re.sub("That's",'That is',text)
    text=re.sub("you've",'you have',text)
    text=re.sub("He's",'He is',text)
    text=re.sub("what's",'what is',text)
    text=re.sub("weren't",'were not',text)
    text=re.sub("we're",'we are',text)
    text=re.sub("hasn't",'has not',text)
    text=re.sub("you'd",'you would',text)
    text=re.sub("shouldn't",'should not',text)
    text=re.sub("let's",'let us',text)
    text=re.sub("they've",'they have',text)
    text=re.sub("You'll",'You will',text)
    text=re.sub("i'm",'i am',text)
    text=re.sub("we've",'we have',text)
    text=re.sub("it's",'it is',text)
    text=re.sub("don't",'do not',text)
    text=re.sub("that´s",'that is',text)
    text=re.sub("I´m",'I am',text)
    text=re.sub("it’s",'it is',text)
    text=re.sub("she´s",'she is',text)
    text=re.sub("he’s'",'he is',text)
    text=re.sub('I’m','I am',text)
    text=re.sub('I’d','I did',text)
    text=re.sub("he’s'",'he is',text)
    text=re.sub('there’s','there is',text)
    
     
    return text
    
dt = df['text'].apply(cleaning)
```
